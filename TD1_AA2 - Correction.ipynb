    - retourne :
        - la liste des indices des noeuds contenants la valeur passée en paramètres, si cette valeur est présentedans l'arbre

```python
def recherche(Arbre, val):
    for i in range(len(Arbre)):
        if Arbre[i][0] == val:
            return Arbre[i][1]
    return []
```
